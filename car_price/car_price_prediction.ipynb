```python
car_models=[]
year_of_productions=[]
engine_volumes=[]
mileages=[]
fuels=[]
gearbox_mechanisms=[] 
steering_wheel_locations=[]
colors=[]
body_types=[]
locations=[] 
price_in_USDs=[]
for page in range(1, 186):
    link='https://www.mashina.kg/search/kia/?currency=2&price_from=&price_to=&page={page}'
    html_text=requests.get(link).content
    soup=BeautifulSoup(html_text, 'lxml')
    cars=soup.find_all('div', class_="list-item list-label")

    for i in cars:
        model=i.find('h2', class_="name").text.replace('Kia', '').replace('\n', '').strip()
        car_models.append(model)
        year_of_production=int(i.find('p', class_="year-miles").find('span').text.replace('\n', '').strip().replace('г.', '').strip())
        year_of_productions.append(year_of_production)
        engine_volume=float( i.find('p', class_="year-miles").text.split(',')[1].replace('л.', '').strip())
        engine_volumes.append(engine_volume)
        mileage=int(i.find('p', class_="volume").text.replace('\n', '').strip().split(",")[-1].replace('км', '').strip().replace(' ',''))
        mileages.append(mileage)
        fuel=i.find('p', class_="body-type").text.replace('\n', '').strip().split(",")[-1].replace(' ', '')
        fuels.append(fuel)
        gearbox_mechanism=i.find('p', class_="year-miles").text.split(',')[-1].replace(' ', '').strip()
        gearbox_mechanisms.append(gearbox_mechanism)
        steering_wheel_location=i.find('p', class_="volume").text.replace('\n', '').strip().split(",")[0].strip()
        steering_wheel_locations.append(steering_wheel_location)
        color=i.find('i', class_="color-icon")['title']
        colors.append(color)
        body_type=i.find('p', class_="body-type").text.replace('\n', '').strip().split(",")[0]
        body_types.append(body_type)
        location=i.find('p', class_="city").contents[0].replace('\n', '').replace(' ', '')
        locations.append(location)
        price_in_USD=int(i.find('div', class_="block price").find('strong').text.replace('$ ', '').replace(' ',''))
        price_in_USDs.append(price_in_USD)

```

```python
all_of_them=pd.DataFrame({
    'car_models':car_models,
    'year_of_productions':year_of_productions,
    'engine_volumes':engine_volumes,
    'mileages':mileages,
    'fuels':fuels,
    'gearbox_mechanisms':gearbox_mechanisms,
    'steering_wheel_locations':steering_wheel_locations,
    'colors':colors,
    'body_types':body_types,
    'locations':locations,
    'price_in_USDs':price_in_USDs
})
```

```python
def compute_cost(x, y, w, b):
   
    m = x.shape[0] 
    cost = 0
    
    for i in range(m):
        f_wb = w * x[i] + b
        cost = cost + (f_wb - y[i])**2
    total_cost = 1 / (2 * m) * cost

    return total_cost
```
